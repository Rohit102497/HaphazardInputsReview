```python
import matplotlib.pyplot as plt
```

```python
Carbon = [0.01, 0.1, 0, 0.32, 0.94, 2.35, 0.18, 1.64, 0.35]
y = ["dummy", "NB3", "FAE", "OLVF", "OCDS", "OVFM", "DynFo", "ORF3V", "Aux-Net", "Aux-Drop"]
y_num = [0, 1, 2, 3, 4, 5, 6, 7, 8]
```

```python
font_size = 12

fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(y_num, Carbon, linestyle='dashed', marker='o',
     markerfacecolor='blue', markersize=6)
# plt.rcParams['font.family'] = 'Times New Roman'

# Annotate
j = 0 # NB_3
ax.annotate(str(Carbon[j]),xy=(y_num[j] - 0.2, Carbon[j] + 0.1), fontsize = font_size) 
j = 1 # FAE
ax.annotate(str(Carbon[j]),xy=(y_num[j] - 0.2, Carbon[j] + 0.1), fontsize = font_size) 
j = 2 # OLVF
ax.annotate(str(Carbon[j]),xy=(y_num[j] - 0.1, Carbon[j] + 0.1), fontsize = font_size) 
j = 3 # OCDS
ax.annotate(str(Carbon[j]),xy=(y_num[j] + 0.17, Carbon[j] - 0.03), fontsize = font_size) 
j = 4 # OVFM
ax.annotate(str(Carbon[j]),xy=(y_num[j] - 0.8, Carbon[j] - 0.05), fontsize = font_size) 
j = 5 # DynFo
ax.annotate(str(Carbon[j]),xy=(y_num[j] + 0.13, Carbon[j] - 0.02), fontsize = font_size) 
j = 6 # ORF3V
ax.annotate(str(Carbon[j]),xy=(y_num[j] - 0.3, Carbon[j] - 0.16), fontsize = font_size) 
j = 7 # Aux-Net
ax.annotate(str(Carbon[j]),xy=(y_num[j] - 0.32, Carbon[j] + 0.1), fontsize = font_size) 
j = 8 # Aux-Drop
ax.annotate(str(Carbon[j]),xy=(y_num[j] - 0.3, Carbon[j] - 0.16), fontsize = font_size) 

ax.set_xticklabels(y,rotation = 45, fontsize = font_size)
ax.tick_params(axis = 'y', labelsize = font_size)

ax.set_xlabel("Models", fontsize = font_size)
ax.set_ylabel("Carbon Emissions (in kg)", fontsize = font_size)

plt.savefig('CarbonEmissions.pdf', bbox_inches = 'tight', dpi = 1200)

# for i,j in zip(y_num, Carbon):
#     ax.annotate(str(j),xy=(i-0.2,j+0.1))
```

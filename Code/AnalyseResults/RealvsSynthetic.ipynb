```python
Real = {'NB3': 83.62,
  'FAE': 84.28,
  'OLVF': 85.44,
  'OCDS': 78.0,
  'OVFM': 73.44,
  'DynFo': 79.11,
  'ORF3V': 83.11,
  'AuxNet': 70.50,
  'AuxDrop': 72.41}

Synthetic = {'NB3': 50.13,
  'FAE': 50.50,
  'OLVF': 57.34,
  'OCDS': 53.37,
  'OVFM': 64.34,
  'DynFo': 62.46,
  'ORF3V': 46.44,
  'AuxNet': 53.39,
  'AuxDrop': 57.95}

real_list = list(Real.values())
synthetic_list = list(Synthetic.values())
y = ["dummy", "NB3", "FAE", "OLVF", "OCDS", "OVFM", "DynFo",	"ORF3V", "Aux-Net", "Aux-Drop"]
y_num = [0, 1, 2, 3, 4, 5, 6, 7, 8]
```

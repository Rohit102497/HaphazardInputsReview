```python
import numpy as np
import result
import pandas as pd
all_result = result.Results()
```

```python
data_list = list(all_result.synthetic_25_mean['NB3'].keys())
prob_list = [0.25, 0.5, 0.75]
model_list = list(all_result.synthetic_25_mean.keys())
```

```python
data_instance = {'wpbc': 198,
 'ionosphere': 351,
 'wdbc': 569,
 'australian': 690,
 'wbc': 699,
 'diabetes_f': 768,
 'german': 959,
 'ipd': 1096,
 'svmguide3': 1243,
 'krvskp': 3196,
 'spambase': 3601,
 'magic04': 19020,
 'a8a': 32561,
 'susy': 1000000,
 'higgs': 2000000}
```

```python
data = []
for d in data_list:
    for p in prob_list:
        for m in model_list:
            if p == 0.25:
                try:
                    val = all_result.synthetic_25_mean[m][d]
                    data.append([d, p, m, val])
                except:
                    print(m, " is not run yet on ", d, " for p = ", p)
            elif p == 0.5:
                try:
                    val = all_result.synthetic_50_mean[m][d]
                    data.append([d, p, m, val])
                except:
                    print(m, " is not run yet on ", d, " for p = ", p)
            elif p == 0.75:
                try:
                    val = all_result.synthetic_75_mean[m][d]
                    data.append([d, p, m, val])
                except:
                    print(m, " is not run yet on ", d, " for p = ", p)
```

```python
data_df = pd.DataFrame(data, columns=['Dataset', 'Prob', 'Model', 'bAcc'])
```
